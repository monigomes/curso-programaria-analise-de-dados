###  Correlação, diferentes funções para dados discretos e contínuos

##### Correlação descreve a relação entre variáveis. Basicamente, ela diz "se" e "como" as variáveis se relacionam entre si.
##### A correlação pode ser positiva quando duas variáveis aumentam juntas. Já a negativa ocorre quando uma variável aumenta e a outra diminui.
##### Correlação de Pearson varia de 1 a -1.

```python
# Correlação de Pearson .corr

correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua

# Apesar da correlação ser positiva, ela está mais próxima do 0 do que do 1,
# então existe a correlação, mas não é tão forte.
# Muito usada em testes de hipóteses
```

```python
from scipy.stats import chi2_contingency
```

```python
# Correlação entre variáveis categorias
# Coeficiente de Cram (0-1) 0-nenhuma associação e 1-associação completa
# min = valor mínimo da tabela -1
# chi2 colocamos o argumento [0], pois não queremos todos os dados, apenas o 1º chi2
# sqrt = raiz quadrada

def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape) -1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])

# Olhando o resultado, a correlação não é muito forte entre ambas. Sugerindo que
# as etnias da área provavelmente possuem nível de ensino semelhantes.
```

### Faça Você Mesme | Correlação

```python
def cramer_coeficiente_2(coluna_1, coluna_2):
  tabela_cruzada2 = np.array(pd.crosstab(coluna_1, coluna_2))
  chi2_2 = chi2_contingency(tabela_cruzada2)[0]
  soma_2 = np.sum(tabela_cruzada2)
  mini_2 = min(tabela_cruzada2.shape) -1
  cramer_2 = np.sqrt(chi2_2/(soma_2*mini_2))
  return cramer_2
```

```python
cramer_coeficiente_2(dados['NIVEL DE ENSINO'], dados['GENERO'])

# Observando o resultado, podemos concluir que existe uma correlação entre
# gênero e nível de ensino, porém não é tão forte. Isso deve ocorrer pelo
# mesmo motivo que no exemplo em aula. Embora a desigualdade de gênero seja nítida,
# é provavel que ambos gêneros que atuam na área de TI tenham o mesmo nível de ensino
# para atuar nessa profissão.
```

```python
lista_estados = list(dados['UF ONDE MORA'].unique())
```

```python
# Use 3 aspas ''' para dizer que tudo após ela é texto, para fechar use três novamente.
# 'AVG' para calcular a média

query = '''SELECT Municipios_Brasileiros.Estado, AVG(Municipio_Status.renda) FROM Municipios_Brasileiros
           INNER JOIN Municipio_Status ON Municipios_Brasileiros.municipio_ID = Municipio_Status.municipio_ID
           WHERE Municipios_Brasileiros.Estado IN ({})
           GROUP BY Municipios_Brasileiros.Estado;'''.format(','.join(['?' for _ in lista_estados]))
```

```python
print(query)
```

```python
# Os pontos de '?' representam os estados presentes na tabela_estados, que serão preenchidos
# por meio do parâmetro 'params'. O 'IN ({})' indica onde será preenchido o '.format'
# Os parâmetros da função são ',' para separar os itens da lista por vírgulas
# '.join([])' para indicar que é uma lista que deve ser atribuído os parâmetros:
# cada estado deve ser preenchido por '?' que será do tamanho da lista de estados: 'for _ in lista_estados'
```

```python
estados_renda = pd.read_sql(query, con=conexao, params=lista_estados)
```

```python
# Renomear coluna de uma tabela
# 'Inplace=True' para atualizar a tabela, aplicando realmente a alteração

dados.rename(columns={'UF ONDE MORA' : 'Estado'}, inplace=True)
```

```python
dados.columns
```

```python
# Juntando tabelas
# Queremos incluir estados_renda por meio (on=) da coluna 'Estados', ficando (how=) com a tabela
# a direita que é a de analise de dados.

dados = dados.merge(estados_renda, on='Estado', how='left')
```

```python
# Correlação Renda e Salário

correlacao_renda_salario = dados['SALARIO'].corr(dados['AVG(Municipio_Status.renda)'])
correlacao_renda_salario
```

```python
# Gráfico Boxplot - "Os impactos de cor/raça/etnia sob o salário"

f, ax = plt.subplots(1, 1, figsize=(10,10))
f.suptitle('Os impactos de cor/raça/etnia sob o salário', fontsize=20, color='blue')

sns.boxplot(x="COR/RACA/ETNIA", y="SALARIO", data=dados, ax=ax, palette='viridis')
ax.set_xlabel("COR/RACA/ETNIA",size = 12,alpha=0.8)
ax.set_ylabel("SALARIO", size = 12, alpha=0.8)
plt.xticks(rotation=60)
plt.grid(True)

plt.tight_layout()

# Neste boxplot, as reflexões que fizemos são novamente conformadas.
# Pessoas amarelas, indígenas e brancas possuem os maiores salários.
# No entanto, as pessoas de cor amarela e branca que possuem valores mais concentrados,
# no terceiro quartil, ou seja, próximos a borda superior. Não é à toa que a
# mediana de ambos está mais próxima do limite inferior, pois os valores pendem para máxima.
# Alguns valores que ultrapassam os limites máximos são observados (pequenos outliers),
# principalmente entre brancos e pardos.
# Pessoas de cor preta são os que recebem menores salários se compararmos as outras
# cores/raças. Há um equilibrio da distribuição entre os quartis e pouquíssimos
# valores que ultrapassam o limite máximo.
# No geral, ficou ainda mais claro que pessoas pretas possuem um salário inferior,
# as outras, e isso já era um dado esperado levando em consideração que a área de
# dados ainda é majoritariamente branca.
```
