## Faça você mesme | Tratando valores faltantes

```python
licit.info()
```

```python
# Verificando onde há NaN

licit.isnull().sum()
```

```python
licit[licit['Data Abertura'].isnull()]
```

```python
licit['Data Abertura'].isnull().sum()
```

```python
# Aqui podemos observar que há 10.648 valores faltantes e apenas 4.688 preenchidos.
# Como os faltantes são expressivamente predominantes, a forma mais segura de lidar com isso
# é optando pela exclusão desses dados, pois eles podem acabar gerando algum tipo de viés.

licit['Data Abertura'].isnull().value_counts()
```

```python
# Apagando os valores NaN

licit.dropna(inplace=True)
licit.shape
```

```python
# Conferindo
# Restaram 4688 linhas no dataset licit

licit['Data Abertura'].isnull().value_counts()
```
